Step 1. Use the file `us_cities_by_population_2010.xlsx` to create a data series. Display the data series. The city/state should be the index.

```python
# Step 1 Solution

import pathlib
import pandas as pd
import numpy as np
import re

cities_2010_file_name = pathlib.Path("data", "us_cities_by_population_2010.xlsx")
cities_2010 = pd.read_excel(cities_2010_file_name, index_col=[1,2])
cities_2010
```

Step 2. Use the file `us_cities_by_population_2020.xlsx` to create a data series. Display the data series. The city/state should be the index.

```python
# Step 2 Solution

cities_2020_file_name = pathlib.Path("data", "us_cities_by_population_2020.xlsx")
cities_2020 = pd.read_excel(cities_2020_file_name, index_col=[1,2])
cities_2020
```

Step 3. Produce a data frame using the two data series. Display the data frame below.

```python
# Step 3 Solution

# option 1

# cities_data = pd.DataFrame({
#     "Rank 1": cities_2010["Rank"],
#     "2010 Actual": cities_2010["2010 Actual"],
#     "Rank 2": cities_2020["Rank"],
#     "2020 Estimate": cities_2020["2020 Estimate"],
# })

# option 2

# cities_data = pd.DataFrame({
#     "2010 Actual": cities_2010["2010 Actual"],
#     "2020 Estimate": cities_2020["2020 Estimate"],
# })

# option 3

# cities_2010.columns = ["Rank 1", "2010 Actual"]
# cities_2020.columns = ["Rank 2", "2020 Estimate"]
# cities_data = cities_2010.join(cities_2020)

# option 4

# cities_data = cities_2010.drop("Rank", axis=1).join(cities_2020.drop(columns=["Rank"]))

# option 5

cities_2010.drop("Rank", axis=1, inplace=True)
cities_2020.drop("Rank", axis=1, inplace=True)

cities_data = cities_2010.join(cities_2020)


cities_data
```

Step 4. Load the 2016 land area for each city from the `us_cities_by_land_area_2016.csv` file. Add the data to data frame.

```python
# Step 4 Solution

cities_2016_land_area_file_name = pathlib.Path("data", "us_cities_by_land_area_2016.csv")

cities_2016_land_area_data = pd.read_csv(
    cities_2016_land_area_file_name,
    skiprows=0,
    index_col=[0,1])

# option 1

# cities_2016_land_area_series = cities_2016_land_area_data["2016 Land Area (sq mi)"]

# land_area_floats = []

# for land_area_val in cities_2016_land_area_series:
#     land_area_floats.append(float(land_area_val.replace(" sq mi", "").replace(",", "")))
    
# land_area_floats[:5]

# cities_data["2016 Land Area (sq mi)"] = land_area_floats

# option 2

cities_data["2016 Land Area (sq mi)"] = cities_2016_land_area_data["2016 Land Area (sq mi)"].apply(
        lambda x: float(x.replace(" sq mi", "").replace(",", ""))
    )

cities_data
```

```python
cities_data
```

Step 5. Calculate a new column for the percent change in population from 2010 to 2020.

```python
# Step 5 Solution

cities_data["Population Change"] = (cities_data["2020 Estimate"] - cities_data["2010 Actual"]) / cities_data["2010 Actual"] * 100

cities_data
```

Step 6. Calculate two columns for population density for both 2010 and 2020.

```python
# Step 6 Solution

cities_data["Population Density 2010"] = cities_data["2010 Actual"] / cities_data["2016 Land Area (sq mi)"]
cities_data["Population Density 2020"] = cities_data["2020 Estimate"] / cities_data["2016 Land Area (sq mi)"]

cities_data
```

Step 7. Save the data frame to a CSV.

```python
# Step 7 Solution

cities_data.to_csv("cities.csv")
```

Step 8. Swap the columns and indexes. Save to a CSV.

```python
# Step 8 Solution

# cities_data.stack().unstack(1).unstack(0)

cities_data2 = cities_data.unstack(0).stack().transpose()

cities_data2.to_csv("cities2.csv")
```

```python
#  Bonus - Clean Up Indexes

# Option 1

# city_index = []
# state_index = []

# annotation = re.compile("\[.*\]")

# for city, state in cities_data.index:
#     city_index.append(re.sub(annotation, "", city.strip()))
#     state_index.append(state.strip())
    
# cities_data.index = pd.MultiIndex.from_arrays([ city_index, state_index ], names=["City", "State"])

# Option 2

annotation = re.compile("\[.*\]")

cities_index_df = pd.DataFrame(list(cities_data.index.values), columns=["City", "State"])

cities_index_df["City"] = cities_index_df["City"].apply(lambda city: re.sub(annotation, "", city.strip()))
cities_index_df["State"] = cities_index_df["State"].str.strip()

cities_data.index = pd.MultiIndex.from_arrays(cities_index_df.values.T, names=cities_index_df.columns)

cities_data.to_csv("cities.csv")
```
